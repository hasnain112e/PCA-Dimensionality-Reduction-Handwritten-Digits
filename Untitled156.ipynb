## Load dataset

### Subtask:
Load the handwritten digits dataset using `sklearn.datasets.load_digits()`.


**Reasoning**:
Import the necessary function, load the dataset, extract data and target, and print their shapes.



```python
from sklearn.datasets import load_digits

digits = load_digits()
X = digits.data
y = digits.target

print(f'Shape of original data (X): {X.shape}')
print(f'Shape of target labels (y): {y.shape}')
```

## Perform pca

### Subtask:
Apply PCA to reduce the dimensionality of the dataset from 64 dimensions to 2 dimensions.


**Reasoning**:
Apply PCA to reduce the dimensionality of the dataset to 2 dimensions and print the shape of the transformed data.



```python
from sklearn.decomposition import PCA

# Instantiate PCA with 2 components
pca = PCA(n_components=2)

# Apply PCA to the data
X_pca = pca.fit_transform(X)

# Print the shape of the transformed data
print(f'Shape of reduced data (X_pca): {X_pca.shape}')
```

## Visualize 2d data

### Subtask:
Create a 2D scatter plot of the reduced data, coloring the points by their original digit labels to visualize clustering.


**Reasoning**:
Generate a scatter plot of the 2D reduced data, coloring the points by their original digit labels to visualize clustering.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
scatter = sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='viridis', legend='full')
plt.title('2D PCA of Digits Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Digit', loc='upper right')
plt.grid(True)
plt.show()
```

## Analyze explained variance

### Subtask:
Calculate and display the explained variance ratio for each component and the cumulative explained variance. Plot the cumulative explained variance and add a line indicating the 95% threshold.


**Reasoning**:
Calculate and display the explained variance ratio and cumulative explained variance, then plot the cumulative explained variance with a 95% threshold line.



```python
# Instantiate PCA with all components to calculate explained variance
pca_full = PCA()
pca_full.fit(X)

# Calculate explained variance ratio for each component
explained_variance_ratio = pca_full.explained_variance_ratio_
print(f'Explained variance ratio for each component:\n{explained_variance_ratio}')

# Calculate cumulative explained variance
cumulative_explained_variance = explained_variance_ratio.cumsum()
print(f'\nCumulative explained variance:\n{cumulative_explained_variance}')

# Plot cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')

# Add the 95% threshold line
plt.axhline(y=0.95, color='r', linestyle='-', label='95% Threshold')

plt.title('Cumulative Explained Variance by Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.legend()
plt.show()
```
